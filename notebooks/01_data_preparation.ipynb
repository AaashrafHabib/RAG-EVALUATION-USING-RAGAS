# 01 - Data Preparation

Ce notebook prépare le corpus et le testset pour l'évaluation RAGAS.

**Corpus choisi** : EU AI Act (Règlement européen sur l'IA)

**Pourquoi ce choix** :
- Sujet récent (2024) → peu de contamination dans les LLMs
- Pertinent pour aivancity (Tech × Business × Éthique)
- Texte structuré avec des articles précis → facilite la vérification factuelle

```python
import sys
sys.path.insert(0, '..')

import json
from pathlib import Path
from src.data_loader import load_corpus_from_texts, load_testset
```

## 1. Chargement du corpus

```python
documents = load_corpus_from_texts('../data/corpus')

print(f"Nombre de documents : {len(documents)}")
print(f"\nDocuments chargés :")
for doc in documents:
    print(f"  - {doc.metadata['source']} ({len(doc.page_content)} caractères)")

total_chars = sum(len(d.page_content) for d in documents)
print(f"\nTotal : {total_chars:,} caractères (~{total_chars//4:,} tokens estimés)")
```

## 2. Chargement et inspection du testset

```python
questions, ground_truths = load_testset('../data/testset/testset.json')

print(f"Nombre de questions : {len(questions)}")
print(f"\n--- Exemples ---")
for i in range(min(5, len(questions))):
    print(f"\nQ{i+1}: {questions[i]}")
    print(f"GT: {ground_truths[i][:100]}...")
```

## 3. Statistiques du dataset

Vérifions la distribution des longueurs de questions et réponses.

```python
import matplotlib.pyplot as plt
import numpy as np

q_lengths = [len(q.split()) for q in questions]
gt_lengths = [len(gt.split()) for gt in ground_truths]

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(q_lengths, bins=10, edgecolor='black', alpha=0.7)
axes[0].set_title('Distribution des longueurs de questions (mots)')
axes[0].set_xlabel('Nombre de mots')
axes[0].set_ylabel('Fréquence')

axes[1].hist(gt_lengths, bins=10, edgecolor='black', alpha=0.7, color='orange')
axes[1].set_title('Distribution des longueurs de ground truths (mots)')
axes[1].set_xlabel('Nombre de mots')
axes[1].set_ylabel('Fréquence')

plt.tight_layout()
plt.savefig('../data/results/dataset_stats.png', dpi=150)
plt.show()

print(f"Questions - Moyenne: {np.mean(q_lengths):.1f} mots, Min: {min(q_lengths)}, Max: {max(q_lengths)}")
print(f"Ground truths - Moyenne: {np.mean(gt_lengths):.1f} mots, Min: {min(gt_lengths)}, Max: {max(gt_lengths)}")
```
